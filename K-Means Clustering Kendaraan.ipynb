```python
# Fungsi untuk mencari outlier
def finding_outlier(df):
  q1 = df.quantile(0.25)
  q3 = df.quantile(0.75)
  iqr = q3 - q1
  df_final = df[(df<(q1-(1.5*iqr))) | (df>(q3+(1.5*iqr)))]
  return df_final
```

```python
# Fungsi untuk memfilter dataframe dari outlier
def remove_outlier(df):
  q1 = df['Premi'].quantile(0.25)
  q3 = df['Premi'].quantile(0.75)
  iqr = q3 - q1
  df_final = df[~((df['Premi']<(q1-(1.5*iqr))) | (df['Premi']>(q3+(1.5*iqr))))]
  return df_final
```

```python
# Mencari jarak antara dua titik
def jarak(data1, data2):
  sum = 0
  for i in range(len(data1)):
    sum += pow((data1[i] - data2[i]), 2) 
  return sqrt(sum)
```

```python
# Menentukan centroid awal
def cent(data,k):
  np.random.seed(5)
  rand = np.random.randint(0,len(data),size=k)
  centroid = [
      #i+1: data.iloc[np.random(data['Premi']), np.random(data['Kanal_Penjualan'])]
      data[i]
      for i in rand
  ]
  return centroid
```

```python
# Mencari centroid terdekat
def near_cent(data, cent):
  nearest = []
  for i in data:
    dist_near = []
    for j in cent:
      dist_near.append(jarak(i,j))
    nearest.append(np.argmin(dist_near))
  return nearest
```

```python
# Mencari centroid baru
def newcent(data, k, cent):
  new_cent = []
  new_data = np.c_[data, cent]
  for i in range(k):
    x = []
    for j in new_data:
      if i == j[-1]:
        x.append(j[:-1])
    df_new = pd.DataFrame(x)
    new_cent.append(df_new.mean(axis=0))
  return new_cent
```

```python
# Mengubah dataframe normal menjadi array
arrnormal = normal.values
arrnormal
```

```python
# Proses K Means Clustering
def kmeans(data,k):
  #random centroid
  centroid = cent(data,k)
  
  while True:
    selisih = 0
    oldCent = centroid
    cluster = near_cent(data, centroid)
    centroid = newcent(data, k, cluster)
    for i in range(len(centroid)):
      
      selisih += jarak(oldCent[i], centroid[i])
    if (selisih == 0):
      break
    
  return cluster, centroid
```

```python
# Proses Evaluasi menggunakan metode SSE
def getSumSquareError(X, assignedCluster, centroids):
  sum = 0
  data = np.c_[X, assignedCluster]
  for i in range(len(centroids)): 
    sumSquare = 0
    for j in data:
      if i == j[-1]:
        sumSquare += jarak(j[:-1], centroids[i])**2
    sum += sumSquare
  return sum
```

```python
cluster, centroid = kmeans(arrnormal, 2)
```

```python
# Visualisasi Elbow Method antara jumlah cluster dengan hasil SSE
wcss = []
k = []
for i in range(1,5):
  cluster, centroid = kmeans(arrnormal, i)
  wcss.append(getSumSquareError(arrnormal, cluster, centroid))
  k.append(i)
```

```python
plt.figure(figsize=(12,8))

plt.plot(k,wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
